Part 1 - Merging and sorting a pandas dataframe

11 lines of python 

```python
import pandas as pd
import numpy as np
```

```python
# Lets make a dummer dataframe
df = pd.DataFrame(np.random.randint(0,100,size=(15, 4)), columns=list('ABCD'))
df.head(3)
```

```python
# Lets make a dummer dataframe, with one of the same columns
df2 = pd.DataFrame(np.random.randint(0,20,size=(15, 4)), columns=list('AZXY'))
df2.head(3)
```

```python
# combine both dataframes into big data, index and sorting need to be turned off
bigdata = pd.concat([df, df2], ignore_index=True, sort=False)
bigdata.head(10)
```

```python
# Sorting the combined data frames by the values in A, or can be done by depth, etc.
foo = bigdata.sort_values(by=['A'])
foo.head(15)
```

```python
# lets export this thing
foo.to_csv('Merged_df.csv')
```
